```python
#Line Plot Basics
import numpy as np
```

```python
x=np.arange(1,11)
```

```python
print(x)
```

```python
y=2*x
```

```python
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
# Adding TItles and Labels
plt.plot(x,y)
plt.title("Line Plot of x and y variable")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
# changing line Aesthetics
plt.plot(x,y,color="orange",linestyle=":",linewidth=3)
plt.title("Line Plot of x and y variable")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color="g",linewidth=3)
plt.plot(x,y2,color="blue",linestyle=":",linewidth=5)
plt.grid(True)
```

```python
# Adding subplots horizontally
import matplotlib.pyplot as plt
plt.subplot(1,3,1)
plt.plot(x,y1,color="g",linewidth=3)
plt.subplot(1,3,2)
plt.plot(x,y1,color="r",linewidth=3)
plt.subplot(1,3,3)
plt.plot(x,y1,color="b",linewidth=3)
plt.show()
```

```python
# Adding subplots vertically
import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.plot(x,y1,color="g",linewidth=3)
plt.subplot(2,1,2)
plt.plot(x,y1,color="r",linewidth=3)
plt.show()
```
